```python
import pandas as pd

# Load the dataset
file_path = 'D:\\twitter_training1.csv'
df = pd.read_csv(file_path)
```

```python
df.columns = ['ID', 'Entity', 'Sentiment', 'Tweet']
df = df.drop(columns=['ID'])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sentiment_distribution = df['Sentiment'].value_counts()


plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_distribution.index, y=sentiment_distribution.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()
```

```python

entity_sentiment_distribution = df.groupby(['Entity', 'Sentiment']).size().unstack().fillna(0)


plt.figure(figsize=(14, 8))
entity_sentiment_distribution.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
plt.title('Sentiment Distribution Across Different Entities')
plt.xlabel('Entity')
plt.ylabel('Frequency')
plt.legend(title='Sentiment')
plt.show()
```
